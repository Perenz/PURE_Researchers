```python
#Get list of titles
pure_titles = list(target_pure_output['Title of the contribution in original language-3'].array)
pure_authors = list(target_pure_output['Contributors-1'].array) 
pure_year = list(target_pure_output['Publication statuses and dates > Date > Year-2'].array)
pure_conference = list(target_pure_output['Title of the host publication in original language-8'].array) 

#dict_keys(['title', 'pub_year', 'author', 'conference', 'pages', 'publisher', 'abstract'])
target_pure_output_list = []
for i in range(target_pure_output.shape[0]):
    add_dict = {
        'title':pure_titles[i],
        'pub_year':pure_year[i],
        'author':pure_authors[i],
        'conference':pure_conference[i]
    }

    target_pure_output_list.append(add_dict)

for f in target_pure_output_list:
    print(f)
```

```python
vectoriser = CountVectorizer()
tpo_bag_of_words = vectoriser.fit_transform(target_pure_output_list).toarray()
print(tpo_bag_of_words)
```

```python
# For each word in target_pure_output, compute similarity with each title in target_publications_title:
#Build SoftTfIdf on google scholar results

soft_tfidf = SoftTfIdf()

#Original W: "Overview of query optimization in XML database systems"
#w_compared = target_pure_output_list[0]

#Typo W: "Overview of query optimization in XML database system"
w_compared = "IMPrECISE: Good-is-good-enough data"
#w_compared = "Point of interest to region of interest conversion"

w_compared_list = w_compared.split()
for w in target_publications_title:
    score = soft_tfidf.get_raw_score(w_compared_list, w.split())
    if score > 0.6:
        print(f'"{w_compared}" and "{w}", score: {score}"')
```

```python
#Loop over google scholar results. If there's no match, add it to the unmatched set
unmatched_titles = set()
treshold = 0.8
soft_tfidf = SoftTfIdf()

for w in target_publications_title:
    #Compute similarity with all word in tarfet_pure_output
    w_list = w.split()
    #Counter of strings with similarity higher than score
    ctr = 0
    for w_c in target_pure_output_list:
        w_c_list = w_c.split()
        if soft_tfidf.get_raw_score(w_c_list, w_list) >= treshold:
            ctr = ctr + 1
            #print(f'Add match {w_c}')
    if ctr == 0:
        #Means there's no match
        #print(f'ctr: {ctr}')
        unmatched_titles.add(w)

```

# List possible warning 

Idea: Create a file (csv?) with all the warning entries

Entry format: Pure data, GScholar data, warning code

- Same name in Pure (D)
- Entry in pure not matching any in GScholar (M)
- Name missing in pure (NM)
- Same name in pure but different:
    - Author (AM)
    - Conference (CM)
    - Pub year (YM)

```python
#use author['filled_publication] for Gscholar, keys; ['title', 'pub_year', 'author', 'conference', 'pages', 'publisher', 'abstract']
#target_pure_output_list for Pure, keys: ['title', 'pub_year', 'author', 'conference']
```

```python
warning_dict = {'D':[], 'M':[], 'NM':[], 'AM':[], 'CM':[], 'YM':[]}
soft_tfidf = SoftTfIdf()
```

```python
#Get codes D
treshold = 1.0
warning_dict['D'] = []
for i1, pure1 in enumerate(target_pure_output_list):
    w1_list = pure1['title'].split()
    for i2, pure2 in enumerate(target_pure_output_list):
        if i1 != i2:
            w2_list = pure2['title'].split()
            score = soft_tfidf.get_raw_score(w1_list, w2_list)
            if score >= treshold:
                #print(score)
                '''
                Create entry to contain:
                w1, w1_data, w2, w2_data, code D, similarity score
                '''
                add_dict = {
                    'title_1':pure1['title'],
                    'pub_year_1':pure1['pub_year'],
                    'author_1':pure1['author'],
                    'conference_1':pure1['conference'],
                    'title_2':pure2['title'],
                    'pub_year_2':pure2['pub_year'],
                    'author_2':pure2['author'],
                    'conference_2':pure2['conference'],
                    'code':'D',
                    'score':score          
                }

                warning_dict['D'].append(add_dict)
```

```python
print(warning_dict)
with open('../warnings/code_D.json', 'w') as outfile:
    json.dump(warning_dict['D'], outfile, indent=4)
```

```python
#Get codes NM Name missing in Pure
treshold=0.8
warning_dict['NM'] = []
for p in author['filled_publications']:
    #Compute similarity with all word in tarfet_pure_output
    w_list = p['title'].split()
    #Counter of strings with similarity higher than score
    ctr = 0
    for pure_o in target_pure_output_list:
        w_c_list = pure_o['title'].split()
        if soft_tfidf.get_raw_score(w_c_list, w_list) >= treshold:
            ctr = ctr + 1
            #print(f'Add match {w_c}')
    if ctr == 0:
        #Means there's no match
        #print(f'ctr: {ctr}')
        add_dict = {
                    'title_1':p['title'],
                    'pub_year_1':p.get('pub_year', ' '),
                    'author_1':p['author'], 
                    'conference_1':p.get('conference', ' '),
                    'title_2':'',
                    'pub_year_2':'',
                    'author_2':'',
                    'conference_2':'',
                    'code':'M',
                    'score':''           
                }
        warning_dict['NM'].append(add_dict)
```

```python
print(warning_dict['NM'])
with open('../warnings/code_NM.json', 'w') as outfile:
    json.dump(warning_dict['NM'], outfile, indent=4)
```

```python
#Get codes M Entry in pure not matching any in GScholar
warning_dict['M'] = []
treshold=0.8
for p in target_pure_output_list:
    #Compute similarity with all word in tarfet_pure_output
    w_list = p['title'].split()
    #Counter of strings with similarity higher than score
    ctr = 0
    for pure_o in author['filled_publications']:
        w_c_list = pure_o['title'].split()
        if soft_tfidf.get_raw_score(w_c_list, w_list) >= treshold:
            ctr = ctr + 1
            #print(f'Add match {w_c}')
    if ctr == 0:
        #Means there's no match
        #print(f'ctr: {ctr}')
        add_dict = {
                    'title_1':p['title'],
                    'pub_year_1':p.get('pub_year', ' '),
                    'author_1':p['author'],
                    'conference_1':p.get('conference', ' '),
                    'title_2':'',
                    'pub_year_2':'',
                    'author_2':'',
                    'conference_2':'',
                    'code':'M',
                    'score':''
                }
        warning_dict['M'].append(add_dict)
```
